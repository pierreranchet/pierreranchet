```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew
from scipy.stats import kurtosis


conn = sqlite3.connect(r"C:\Users\ranch\Desktop\Skema\S2\Python 2\projectss_db.db")
df = pd.read_sql_query("select * from HISTO", conn)
df['DATE'] =  pd.to_datetime(df['DATE'], format='%Y-%m-%d')
df.head(60)

df = df.set_index('DATE')
df.drop(['ID'], axis=1, inplace=True)


SX5E = df['VALUE'][df['TICKER'] == 'SX5E Index']
SX5E_logret = np.log(SX5E) - np.log(SX5E.shift(1))
SX5E_logret = SX5E_logret[1:2506]
SX5E_logret = SX5E_logret.resample('MS').sum()

SPX = df['VALUE'][df['TICKER'] == 'SPX Index']
print(len(SPX))
SPX_logret = np.log(SPX) - np.log(SPX.shift(1))
print(len(SPX_logret))
SPX_logret = SPX_logret[1:2456]
SPX_logret = SPX_logret.resample('MS').sum()

GOLD = df['VALUE'][df['TICKER'] == 'XAUUSD Curncy']
GOLD_logret = np.log(GOLD) - np.log(GOLD.shift(1))
GOLD_logret = GOLD_logret[1:2540]
GOLD_logret = GOLD_logret.resample('MS').sum()

WTI = df['VALUE'][df['TICKER'] == 'CL1 Comdty']
WTI_logret = np.log(WTI) - np.log(WTI.shift(1))
WTI_logret = WTI_logret[1:2459]
WTI_logret = WTI_logret.resample('MS').sum()

print('SX5E Index log returns :  \n', 'Mean : ', round(SX5E_logret.mean()*100, 4),
      '%\n', 'Variance : ', round(SX5E_logret.var()*100, 4), '%\n', 'Skewness : ', round(skew(SX5E_logret), 4),'\n',
    'Kurtosis : ', round(kurtosis(SX5E_logret), 4))

print('\nSPX Index log returns :  \n', 'Mean : ', round(SPX_logret.mean()*100, 4),
      '%\n', 'Variance : ', round(SPX_logret.var()*100, 4), '%\n', 'Skewness : ', round(skew(SPX_logret), 4),'\n',
    'Kurtosis : ', round(kurtosis(SPX_logret), 4))

print('\nGOLD log returns :  \n', 'Mean : ', round(GOLD_logret.mean()*100, 4),
      '%\n', 'Variance : ', round(GOLD_logret.var()*100, 4), '%\n', 'Skewness : ', round(skew(GOLD_logret), 4),'\n',
    'Kurtosis : ', round(kurtosis(GOLD_logret), 4))

print('\nWTI log returns :  \n', 'Mean : ', round(WTI_logret.mean()*100, 4),
      '%\n', 'Variance : ', round(WTI_logret.var()*100, 4), '%\n', 'Skewness : ', round(skew(WTI_logret), 4),'\n',
    'Kurtosis : ', round(kurtosis(WTI_logret), 4))
```
